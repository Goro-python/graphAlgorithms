```python
import collections

import heapq;

class Graph(object):
    
    def __init__(self,nodeList):
        
        self.nodeList=nodeList;
        self.edgeList=[];
        self.dic=collections.defaultdict(list);
        for node in self.nodeList:
            self.dic[node]=[];
            
        self.edgeList=[];
        
        
        self.distMatrix=[[float('inf') for node in self.nodeList]for node in self.nodeList];
        
        
        self.mstPath =[];
        
        
        self.permutations=[];
            
    def addEdge(self,dest,source,cost):  # cost 1 for source to dest  #cost2 for dest tyo source
    
        self.dic[source].append((dest,cost));
        self.dic[dest].append((source,cost));
        
        self.edgeList.append((source,dest,cost));
        self.edgeList.append((dest,source,cost));
        
        
        self.distMatrix[source][source]=0;
        self.distMatrix[dest][dest]=0;
        self.distMatrix[source][dest]=cost;
        self.distMatrix[dest][source]=cost;
        
        
    def addEdgeDir(self,source,dest,cost):
        
        self.distMatrix[source][dest]=cost;
        self.distMatrix[source][source]=0;
        self.distMatrix[dest][dest]=0;
        
        
        
    def dijkstra(self,start):  # it can handle negative edges and cycle
        
        
        predec=[-1]*len(self.nodeList);
        destCost=[float('inf')]*len(self.nodeList);
        destCost[start]=0;
        utility=[[float('inf'),node] for node in self.nodeList];        
        utility[start]=[0,start];
        
        heapq.heapify(utility); #O(V+logV)
        
        currentSet=[False]*len(self.nodeList);
        
        while utility:
            
            #print(utility);
            (cost,node) = heapq.heappop(utility);
            
            currentSet[node]=True;
            
           
            #print(destCost);
            
            
            for (ne,cost) in self.dic[node]:
                
            
                
                if currentSet[ne] is False:
                    
                    if destCost[ne]>destCost[node]+cost:
                        
                        destCost[ne]=destCost[node]+cost;
                        predec[ne]=node;
                        heapq.heappush(utility,[destCost[ne],ne]);
                        
                        
        print(destCost);
        print(predec)
        
        
    def bellmanFord(self,start):
        
        destCost=[float('inf')]*len(self.nodeList);
        predec=[-1]*len(self.nodeList);
        
        destCost[start]=0;
        
        for times in range(1,len(self.nodeList)):
            print(destCost)
            for s,d,cost in self.edgeList:
        
                if destCost[d]>destCost[s]+cost:
                    destCost[d]=destCost[s]+cost;
                    predec[d]=s;
                    
            
                
                
        
        #checking for negative dge cycle
        
        for s,d,cost in self.edgeList:
            
            if destCost[d]>destCost[s]+cost:
                
                print('negative cycle detected');
                return;
            
            
        print(destCost);
        
        print(predec);
        
    def floydWarshall(self):
        
        for passNode in self.nodeList:
            for src in self.nodeList:
                for dest in self.nodeList:    
                    
                
                    if self.distMatrix[src][dest]>self.distMatrix[src][passNode]+self.distMatrix[passNode][dest]:

                            
                        self.distMatrix[src][dest]=self.distMatrix[src][passNode]+self.distMatrix[passNode][dest];
                            
                            
                            
        print(self.distMatrix);
        
    def Prims(self):   #MST
        
        
        dist=[float('inf') for node in self.nodeList];
        mstSet=[False]*len(self.nodeList);
        predec=[-1]*len(self.nodeList);
        
        dist[0]=0;
        
        utility = [(dist[node],node ) for node in self.nodeList];
        
        heapq.heapify(utility);
        
        
        while utility:

            util,node  =  heapq.heappop(utility);
            
            mstSet[node]=True;
            
            self.mstPath.append(node);
            for ne,cost in self.dic[node]:
                
                if mstSet[ne] is False:
                    
                    if dist[ne]>cost:
                        dist[ne]=cost;
                        predec[ne]=node;
                        heapq.heappush(utility,(dist[ne],ne));
                    
                    
            if all(mstSet):
                break;
                
                
                
        print(dist);
        
        
    def tspNaive(self):
        

        def helper(arr,cur,parent,dist):
    
            if not arr:
                path=cur.copy();
                print('path: ',path)
                self.permutations.append((path+[path[0]],dist+self.distMatrix[parent][path[0]]));
                return;
            else:
            
                for i,num in enumerate(arr):
                    cur.append(num);

                    if parent!=-1:
                        dist+=self.distMatrix[parent][num];
                    
                    helper(arr[:i]+arr[i+1:],cur,num,dist);
                    
                    if parent!=-1:
                        dist-=self.distMatrix[parent][num];
                        
                    cur.pop();
                    
                    
                return;
                
        
        self.permutations=[];
        
        helper(self.nodeList,[],-1,0);
        
        
        #print(self.permutations);
        
        
        self.permutations.sort(key = lambda perm: perm[1]);
        
        
        print(self.permutations);
        
        
        
        
        

                    
            
        
        
        
        
        
        
        
        
        
        
    
                    
        
        
        
        
        

        
        
        
        
        
        
        
        
    
                        
                    
            
        
        
        
        
        
        
                
                
            
            
            
        
        
        
                
                
    
                
                
                
            
            
            
            
            
            
            
            
            
            
            
            
        
        
        
        
        
        
        

        
        
#testcase :


nodeList=[0,1,2,3]

G=Graph(nodeList);


G.addEdgeDir(0,1,10);
G.addEdgeDir(0,2,15);
G.addEdgeDir(0,3,20);
G.addEdgeDir(1,0,10);
G.addEdgeDir(1,2,35);
G.addEdgeDir(1,3,25);
G.addEdgeDir(2,0,15);
G.addEdgeDir(2,1,35);
G.addEdgeDir(2,3,30);
G.addEdgeDir(3,0,20);
G.addEdgeDir(3,1,25);
G.addEdgeDir(3,2,30);


#.bellmanFord(0);



# { 0, 10, 15, 20 }, 
# { 10, 0, 35, 25 }, 
# { 15, 35, 0, 30 }, 
# { 20, 25, 30, 0 } 
        
G.tspNaive();






        
        
        
        
        
        

        
            
            
    
        
            
            
        
```
